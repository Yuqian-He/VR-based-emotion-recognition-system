```python
import opensmile
import os
import pandas as pd
import sounddevice as sd
```

```python
import IPython.display as ipd
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)
# feature_set=["ComParE_2016","GeMAPSv01a","GeMAPSv01b","eGeMAPSv01a","eGeMAPSv01b","eGeMAPSv02"]

len(smile.feature_names)
```

```python
duration = 5  # seconds
fs = 44100
# Start recording
recording = sd.rec(int(duration * fs), samplerate=fs, channels=1)
print("Recording started.")
sd.wait()  # Wait for the recording to finish
print("Recording finished.")

feature_df = smile.process_signal(signal =recording.T,sampling_rate=fs)
feature = feature_df.to_numpy().squeeze(0)

filename = "RecordingData.txt"

# Open the file for writing
with open(filename, "a") as file:
    # Write some text to the file
    for i in feature:
        file.write(str(i)+",")
    file.write("\n")

ipd.Audio(data = recording.T,rate = fs)
```
